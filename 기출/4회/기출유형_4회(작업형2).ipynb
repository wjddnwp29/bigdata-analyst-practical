## Q. [마케팅] 자동차 시장 세분화
- 자동차 회사는 새로운 전략을 수립하기 위해 4개의 시장으로 세분화했습니다.
- 기존 고객 분류 자료를 바탕으로 신규 고객이 어떤 분류에 속할지 예측해주세요!


- 예측할 값(y): "Segmentation" (1,2,3,4)
- 평가: Macro f1-score
- data: train.csv, test.csv
- 제출 형식:
~~~
ID,Segmentation
458989,1
458994,2
459000,3
459003,4
~~~

### 노트북 구분
- basic: 수치형 데이터만 활용 -> 학습 및 test데이터 예측
- intermediate: 범주형 데이터도 활용 -> 학습 및 test데이터 예측
- advanced: 학습 및 교차 검증(모델 평가) -> 하이퍼파라미터 튜닝 -> test데이터 예측

```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 불러오기
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/4_2/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/4_2/test.csv")
```

```python
train.shape, test.shape
```

```python
target = train.pop("Segmentation")
```

```python
ID = test.pop("ID")
```

```python
test.info()
```

```python
train = train.drop("ID",axis=1)
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l = ["Gender","Ever_Married","Graduated","Profession","Spending_Score","Family_Size","Var_1"]
for i in l:
    le = LabelEncoder()
    train[i] = le.fit_transform(train[i])
    test[i] = le.transform(test[i])
train.head()  
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_val,y_tr,y_val = train_test_split(
    train,
    target,
    test_size = 0.2,
    random_state = 0
)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 42, max_depth = 200,n_estimators = 300)
model = rf.fit(x_tr,y_tr)
pred = model.predict(x_val)
pred
```

```python
from sklearn.metrics import f1_score
ans = f1_score(y_val,pred, average = "macro")
ans
```

```python
ans = model.predict(test)
```

```python
import pandas as pd
result = pd.DataFrame({
    "ID" : ID,
    "Segmentation" : ans
})
result
```

```python
result.to_csv("result.csv",index=False)
```
