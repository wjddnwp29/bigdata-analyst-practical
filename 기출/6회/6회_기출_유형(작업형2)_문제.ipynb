### 답안 제출 참고
- pd.read_csv('result.csv') 로 제출 코드 확인

```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_2/energy_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_2/energy_test.csv")
train.shape, test.shape
```

```python
train.head()
target = train.pop("Heat_Load")
```

```python
train = train.drop("Roof",axis=1)
test = test.drop("Roof",axis=1)
```

```python
test.head()
```

```python
tmp = ["Height","Orient"]
from sklearn.preprocessing import LabelEncoder
for i in tmp:
    le = LabelEncoder()
    train[i] = le.fit_transform(train[i])
    test[i] = le.transform(test[i])
train.head()
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_val,y_tr,y_val = train_test_split(
    train,
    target,
    test_size = 0.1,
    random_state = 0
)
x_tr.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
model = rf.fit(x_tr,y_tr)
pred = model.predict(x_val)
pred

import lightgbm as lgb
lg = lgb.LGBMClassifier(random_state=0)
model = lg.fit(x_tr,y_tr)
pred1 = model.predict(x_val)
pred1
```

```python
from sklearn.metrics import f1_score
ans = f1_score(y_val,pred1,average = "macro")
ans
```

```python
pred = model.predict(test)
import pandas as pd
a = pd.DataFrame({
    "pred" : pred
})
a.to_csv("result.csv",index=False)
```
