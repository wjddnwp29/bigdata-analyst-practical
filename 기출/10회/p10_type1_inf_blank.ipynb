### 문제2
제공된 데이터(cafe_sales.csv)는 2024년 4월부터 2025년 12월까지의 카페 메뉴 판매 이력이다. 소문항을 각각 구한 다음 [제출 형식]에 맞춰 답안 페이지에 입력하시오.

제공 데이터
- order_date : 주문 날짜 (예 2024.05.24 13:22)
- category : 음료/디저트 종류 (예 coffee, tea …)
- item : 메뉴명
- price : 판매 금액

소문항

2-1. 연-월별 총 매출액을 계산하여 큰 순서대로 정렬했을 때, 2번째로 큰 매출액을 입력하시오.

2-2. 연-월별 총 매출액이 4번째로 큰 연-월을 찾으시오. 해당 연-월의 데이터에서 카테고리별 매출 합계를 계산하고, 그중 최댓값을 입력하시오.

[제출 형식] 정수로 작성


```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/10_1/cafe_sales.csv")
df.info()
df.head()
```

```python
# 2-1
# order_date → 연-월 변환
df["order_date"] = pd.to_datetime(df["order_date"])
# 2-1
# 1. 연-월 단위로 그룹화하여 각 월의 총 매출액(price 합계)을 계산
df["년도"] = df["order_date"].dt.year
df["월"] = df["order_date"].dt.month
# 정답: 328741
```

```python
# 2. 연-월별 총매출액 중 상위 2개만 추출
df.groupby(["년도","월"])["price"].sum().sort_values(ascending = False)
```

```python
# 2-2

# 1. 연-월별 총매출액 중 상위 4개만 추출

# 2. 4번째로 큰 연-월에 해당하는 데이터만 추출
cond1 = df["년도"] == 2024
cond2 = df["월"] == 10
cond = (cond1&cond2)
tmp = df[cond]
tmp.groupby("category")["price"].sum().max()
# 3. 해당 연-월의 카테고리별 매출 합계 중 최댓값 계산

# 정답: 104660
```
