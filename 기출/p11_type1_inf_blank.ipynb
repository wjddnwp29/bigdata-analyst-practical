## 문제 1

제공된 데이터(`greenhouse_gas.csv`)는 2001년부터 2021년까지 30개 국가의 연도별 온실가스 배출량 데이터입니다. 아래 소문제를 각각 구하시오.

### 문제 1-1

2001년부터 2021년까지 각 연도별로 온실가스 배출량이 가장 많은 국가를 찾고, 그 중 1위를 가장 많이 차지한 국가의 1위 횟수를 정수로 구하시오.

### 문제 1-2

다음 공식으로 각 국가의 CAGR(연평균 성장률)을 계산하고, 그 중 최댓값을 소수점 셋째 자리까지 구하시오.
- CAGR = (마지막 값 / 처음 값)<sup>1/기간</sup> - 1
- CAGR = (2021년 배출량 / 2001년 배출량)<sup>1/20</sup> - 1

**제공 데이터 (`greenhouse_gas.csv`)**

- `Country` : 국가명 (Brazil, China, USA, India 등 30개국)
- `2001` : 2001년 온실가스 배출량
- `2002` : 2002년 온실가스 배출량
- `2003` : 2003년 온실가스 배출량
- ...
- `2021` : 2021년 온실가스 배출량
