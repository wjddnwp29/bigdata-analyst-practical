## 문제

제공된 학습용 데이터(`train.csv`)는 온라인 게임 플레이어의 활동 데이터이다. 학습용 데이터를 활용하여 유저 등급(`User_Level`)을 예측하는 모형을 개발하고, 모형을 평가용 데이터(`test.csv`)에 적용하여 유저 등급을 예측하시오. 예측 결과는 아래의 **[유의 사항]**과 **[제출 형식]**을 준수하여, CSV 파일(`result.csv`)을 생성하는 코드를 제출하시오.

## 유의 사항

1. 예측 결과는 **F1 Score(macro)** 평가 지표에 따라 평가함  
2. 타깃 변수(`User_Level`)는 **0, 1, 2 총 3개의 클래스로 구성됨**
   - `0`: Low
   - `1`: Medium
   - `2`: High

## 제출 형식

1. CSV 파일명: `result.csv`  
2. 예측 유저 등급 컬럼명: `pred`  
3. 제출 컬럼 개수: `pred` 컬럼 1개  
4. 평가용 데이터 개수와 예측 결과 데이터 개수 일치: **903개**

## 제공 데이터

````python
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p4/11/game_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p4/11/game_test.csv")

# 데이터 전처리
target = train.pop('User_Level')

# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

# 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)

# macro F1 score
from sklearn.metrics import f1_score
print("랜덤포레스트")
print(f1_score(y_val, pred, average='macro'))

# LightGBM
import lightgbm as lgb
lg = lgb.LGBMClassifier(random_state=0, verbose=-1)
lg.fit(X_tr, y_tr)
pred = lg.predict(X_val)
print("lightgbm")
print(f1_score(y_val, pred, average='macro'))

# 최종 제출 파일 (lightGBM)
pred = lg.predict(test)
result = pd.DataFrame({
    'pred':pred
})
result.to_csv("result.csv", index=False)
````

## 심화

````python
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p4/11/game_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p4/11/game_test.csv")

# 데이터 전처리
target = train.pop('User_Level')

# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0, stratify=target)

# 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0, class_weight='balanced')
rf.fit(X_tr, y_tr)
pred = rf.predict(X_val)

# macro F1 score
from sklearn.metrics import f1_score
print("랜덤포레스트")
print(f1_score(y_val, pred, average='macro'))

# LightGBM
import lightgbm as lgb
lg = lgb.LGBMClassifier(random_state=0, verbose=-1, class_weight='balanced')

lg.fit(X_tr, y_tr)
pred = lg.predict(X_val)
print("lightgbm")
print(f1_score(y_val, pred, average='macro'))

# 최종 제출 파일 (lightGBM)
pred = lg.predict(test)
result = pd.DataFrame({
    'pred':pred
})
result.to_csv("result.csv", index=False)
````

````python
# LightGBM - 전체 데이터로 학습
lg = lgb.LGBMClassifier(random_state=0, verbose=-1, class_weight='balanced')
lg.fit(train, target)

# 최종 제출 파일 (lightGBM)
pred = lg.predict(test)
result = pd.DataFrame({
    'pred': pred
})
result.to_csv("result.csv", index=False)
````
