## 문제 1

제공된 데이터(`elderly_health.csv`)는 보건소에서 수집한 건강 검진 데이터로,  
나이, 당뇨병 이력, 신체 활동 여부, 공복 혈당 등에 따른 노령층 여부(predicted)에 관한 자료이다.  

이 데이터를  
- **Train 데이터**: id 1 ~ 2000  
- **Test 데이터**: id 2001 ~ 2227  

로 나누어 로지스틱 회귀분석(Logistic Regression)을 수행하고, 제시된 소문항에 답하시오.

---

### 제공 데이터 (`elderly_health.csv`)

### 종속 변수 (타겟 변수)
- **predicted** : 노령층 여부  
  - 0 = 비노령층  
  - 1 = 노령층  

### 독립 변수 (설명 변수)
- **id** : 고객 번호  
- **age** : 나이  
- **diabetic** : 당뇨병 이력  
- **activity** : 신체 활동 여부  
- **glus_fast** : 공복 혈당  
- **bmi** : 체질량지수  
- **blood_pressure** : 혈압  


---

## 문제 1-1

Train 데이터에서 id 변수를 제외한 모든 변수를 활용하여 로지스틱 회귀분석(Logistic Regression)을 수행하시오.

`diabetic`이 'no'인 사람 대비 'yes'인 사람의 오즈비(Odds Ratio)를 구하시오.

- 반올림하여 소수 둘째 자리까지 작성

---

## 문제 1-2

위 모델에서 test 데이터 중 노령층일 확률이 가장 높은 사람의 `glus_fast`(공복 혈당) 값을 구하시오.

- 정수로 작성

---

## 문제 1-3

위 모델로 test 데이터를 예측한 후,  
확률이 0.2 초과인 사람을 노령층으로 분류했을 때 민감도(Sensitivity)를 구하시오.

- 반올림하여 소수 둘째 자리까지 작성

```python
# 1-1
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/elderly_health.csv")
train,test = df.iloc[:2000], df.iloc[2000:]
train.shape,test.shape
train = train.drop("id",axis=1)
```

```python
from statsmodels.formula.api import logit
model = logit("predicted ~ age+diabetic+activity+glus_fast+bmi+blood_pressure",train).fit()
# print(model.summary())
# `diabetic`이 'no'인 사람 대비 'yes'인 사람의 오즈비(Odds Ratio)를 구하시오.
import numpy as np
ans = np.exp(3.0138 )
ans
```

```python
# 1-2
## 문제 1-2
# 위 모델에서 test 데이터 중 노령층일 확률이 가장 높은 사람의 `glus_fast`(공복 혈당) 값을 구하시오.
# - 정수로 작성
pred = model.predict(test)
a = pred.idxmax()
a
```

```python
test.loc[a]
```

```python
# # 1-3
pred_labels = (pred > 0.2).astype(int)
actual_labels = test["predicted"]
from sklearn.metrics import recall_score
ans = recall_score(actual_labels,pred_labels)
ans
```

- sales : 매출액

## 문제 2-1

홍보 전후 매출 평균이 35,000원과 같은지 단일 t검정(t-test)을 수행하고 p-value를 구하시오.

- 반올림하여 소수 셋째 자리까지 작성

---

## 문제 2-2

매출액(sales)과 다른 모든 변수들 간의 피어슨 상관계수(Pearson Correlation Coefficient)를 구하고, 그 중 상관관계가 가장 강한(절댓값이 가장 큰) 상관계수 값을 구하시오.

- 반올림하여 소수 셋째 자리까지 작성

---

## 문제 2-3

매출액(sales)을 종속변수, 나머지 모든 변수를 독립변수로 하여 회귀분석을 수행하시오.

절편을 포함한 모든 변수 중에서 p-value가 가장 작은 변수(통계적으로 가장 유의한 변수)의 회귀계수를 구하시오.

- 반올림하여 소수 셋째 자리까지 작성

```python
# 2-1
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bae-v3/main/part4/ch11/promotion_data.csv")
df.head()
```

```python
import pandas as pd
from scipy.stats import ttest_1samp
# 2. 단일 t-검정 수행 (sales 컬럼의 평균이 35,000과 같은지)
result = ttest_1samp(df['sales'], 35000)

# 3. p-value 추출 및 반올림 (소수 셋째 자리까지)
p_value = result.pvalue
ans = round(p_value, 3)

# 최종 정답 출력
print(ans)
```

```python
# 2-2
"""
---

## 문제 2-2

매출액(sales)과 다른 모든 변수들 간의 피어슨 상관계수(Pearson Correlation Coefficient)를 구하고, 그 중 상관관계가 가장 강한(절댓값이 가장 큰) 상관계수 값을 구하시오.

- 반올림하여 소수 셋째 자리까지 작성

---
"""
a = df.corr(method="pearson")
a
# 0.575396
```

```python
# 2-3
'''
## 문제 2-3

매출액(sales)을 종속변수, 나머지 모든 변수를 독립변수로 하여 회귀분석을 수행하시오.

절편을 포함한 모든 변수 중에서 p-value가 가장 작은 변수(통계적으로 가장 유의한 변수)의 회귀계수를 구하시오.

- 반올림하여 소수 셋째 자리까지 작성
'''
from statsmodels.formula.api import ols
model = ols("sales ~ visit_count + page_time + ad_clicks + promotion_budget",df).fit()
model.summary()
# 130.7392
```
