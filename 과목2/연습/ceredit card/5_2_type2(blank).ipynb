## 데이터 전처리 & 피처엔지니어링

```python
train.head()
```

```python
train.nunique()
print(train.shape)
# name, host_id, host_name, last_review
cols = ['name', 'host_id', 'host_name', 'last_review']
train = train.drop(cols,axis=1)
test = test.drop(cols,axis=1)
print(train.shape)
```

```python
# reviews_per_month -> 0
train['reviews_per_month'] = train['reviews_per_month'].fillna(0)
test['reviews_per_month'] = test['reviews_per_month'].fillna(0)
train.isnull().sum()
```

```python
train = train.drop('id',axis=1)
test_id = test.pop('id')
test.head()
```

```python
train.info()
```

```python
cols = ['neighbourhood_group','neighbourhood','room_type']
from sklearn.preprocessing import LabelEncoder

for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
train[cols]
```

## 검증 데이터 분리

```python
from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val = train_test_split(
    train.drop('price',axis=1),
    train['price'],
    test_size = 0.15,
    random_state = 2022)
x_tr
```

## 모델 & 평가

```python
# 평가
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
def rmse(y_test, y_pred): #RMSE
    return np.sqrt(mean_squared_error(y_test, y_pred))

def rmsle(y_test, y_pred): #RMSLE
    return np.sqrt(np.mean(np.power(np.log1p(y_test) - np.log1p(y_pred), 2)))

def mape(y_test, y_pred): #MAPE
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
from sklearn.ensemble import RandomForestRegressor
rg = RandomForestRegressor()
rg.fit(x_tr,y_tr)
pred=rg.predict(x_val)
```
