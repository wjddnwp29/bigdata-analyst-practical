* 실기시험에서는 무조건 표본데이터
* 표본평균구하라하면 그냥 평균구하면됨.

* 귀무가설 : 0가설 기존에 알려진 가설, 효과나 차이가 없음
* 대립가설 : 연구자가 입증 하려는 사실, 효과나 차이가 있음

* 검정결과
* 검정통계량 : 주어진 데이터와 귀무가설 간의 차이르 통계적으로 나타내는 값
* p-value (유의수준 0.05)
* 유의수준보다 작으면 귀무가설을 기각하고, 대립가설을 채택한다.
* 검정 통계량 및 P-value(유의확률) 계산

* 단일표본검정 - 모집단 1개
* 대응표본검정 - 모집단 2개 신약 효과(전후) 검정
* 독립표본검정 - 모집단 2개

ㅊ적합도 검정
* scipy.stats.chisquare(관찰된 빈도 리스트, 기대 빈도 리스트)
* 빈도로 넣어주는게 중요
* 관찰도수와 기대도수의 차이

* 독립성 검정
* scipy.stats.chi2_contingency(table, correction=True)
* 두 변수가 서로 독립적인지(연관성이 있는지) 확인
* 교차표를 테이블로 만들어야함.
* scipy.stats.chi2_contingency(table, corretion = True)
* correction = 연속성보정(기본값 True)
* 동질성 검정도 동일

* 상관관계 - 두 변수 간의 선형적인 관계
