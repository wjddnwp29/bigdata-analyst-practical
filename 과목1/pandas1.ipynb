## Section1. 데이터프레임과 시리즈

```python
import pandas as pd
# Series도 표로 렌더링되게 만들기
pd.Series._repr_html_ = lambda self: self.to_frame()._repr_html_()
```

```python
# 시리즈 만들기
menu = pd.Series(['비빔밥','김치찌개','된장찌개'])
menu
```

```python
# 시리즈 만들기
price = pd.Series([10000,9000,8000])
price
```

```python
# 데이터 프레임 만들기
pd.DataFrame({
    "메뉴" : menu,
    "가격" : price
})
```

```python
df = pd.DataFrame({
    "메뉴" : ['비빔밥', '김치찌개','된장찌개'],
    "가격" : [10000,9000,8000],
    "원산지" : ["국내산","국내산","국내산"]
})
df
```

```python
# 시리즈 선택
df['메뉴']
```

```python
# 데이터 프레임으로 선택하는 것 [[]]
df[['메뉴','가격']]
```

```python
df
```

```python
# 데이터프레임 선택(메뉴, 가격)
cols = ['메뉴', '가격']
df[cols]
```

```python
print("df            :",type(df))
print("df            :",type(df['가격']))
print("df            :",type(df[['가격']]))
```

## Section2. 데이터 저장 및 불러오기
