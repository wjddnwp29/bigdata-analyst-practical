## Section4. 자료형 변환

```python
import pandas as pd
# 데이터 준비
data = {
    "메뉴":['아메리카노', '카페라떼', '카페모카', '카푸치노', '에스프레소', '밀크티', '녹차'],
    "가격":[4500.0, 5000.0, 5500.0, 5000.0, 4000.0, 5900.0, 5300.0],
    "칼로리":['10', '110', '250', '110', '20', '210', '0'],
}
```

```python
df = pd.DataFrame(data)
df.info()
```

```python
# 자료형 변환 / astype / float -> int
df['가격'] = df['가격'].astype('int')
df.info()
```

```python
# 자료형 변환 / astype / object -> float
df['칼로리'] = df['칼로리'].astype('int')
df.info()
```

## Section5. 새로운 컬럼 추가

```python
df = pd.read_csv('cafe.csv')
```

```python
df.head(2)
```

```python
# 새로운 컬럼 추가(값은 0)
df['new'] = 0
df.head(2)
```

```python
# 할인가 컬럼 추가
discount = 0.2
df['할인가'] = df['가격'] * (1-discount)
df.head(2)
```

## Section6. 데이터 삭제(칼럼, 행)

```python
# 행 삭제
# axis = 0 ==> 행을 선택하겠다.
df = df.drop(1, axis=0)
df.head(3)
```

```python
# 칼럼 삭제
# axis = 1 : 열방향삭제
df = df.drop('칼로리',axis=1)
df.head(3)
```

## Section7. 인덱싱/슬라이싱

```python
df = pd.read_csv('cafe.csv')
df
```

```python
# 인덱싱 (행 전체)
# 아메리카노
df.loc[0]
```

```python
# 카페라테 가격
df.loc[1,"가격"]
```

```python
# 슬라이싱 (컬럼 전체)
# 가격
df.loc[:,"가격"]
```

```python
# 슬라이싱
# 카페모카 메뉴와 칼로리
df.loc[2, ['메뉴','칼로리']]
```

```python
# 슬라이싱
df.loc[1:3,'메뉴':'가격']
```
