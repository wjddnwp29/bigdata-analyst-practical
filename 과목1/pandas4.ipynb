## Section14. 문자열

```python
import pandas as pd
df = pd.DataFrame({"A":['데이터 분석','기본 학습서','퇴근 후 열공'],
                   "B":[10,20,30],
                   "C":['ab cd','AB CD','ab cd ']
                   })
df
```

```python
# 문자의 일부는 .replace로 X
df['A'] = df['A'].replace("분석", "시각화")
df
```

```python
df['A'] = df['A'].str.replace('분석','시각화')
df
```

```python
# 에러
# df['B'] = df['B'].str.replace(10,20)
# df
```

```python
# 문자열 분리
df['A'].str.split()
```

```python
# 앞쪽 or 뒤쪽
df['A'].str.split()[0]
```

```python
df['D'] = df['A'].str.split().str[0]
df
```

```python
# 문자 검색
df['A'].str.contains('기본')
```

```python
df['기본포함유무'] = df['A'].str.contains('기본')
df
```

```python
# 단어의 일부로는 못찾음
menu = pd.Series(['맛난버거 세트','매운 치킨버거','더블 치즈버거'])
menu.isin(['맛난버거 세트','더블 치즈버거'])
```

```python
# 단어의 일부로 찾을 수 있음
menu.str.contains('세트')
```

## Section15. 내장함수

```python
df = pd.read_csv("cafe4.csv")
df.head(2)
```

```python
# 행의 수
print(len(df))
print(df.shape[0])
```

```python
cond = df['가격'] > 5000

# sum으로 조건에 맞는 개수 구하기
print(sum(cond))

# len으로 조건에 맞는 개수 구하기 
print(len(df[cond]))
```

```python
df
```

```python
# 열 별 합계
df.sum(numeric_only=True)
```

```python
# 행 별 합계
df.sum(axis = 1, numeric_only=True)
```

```python
print("최대값: ",df['가격'].max())
print("최소값: ",df['가격'].min())
print("평균값: ",df['가격'].mean())
print("중앙값: ",df['가격'].median())
print("합계: ",df['가격'].sum())
print("표준편차: ",df['가격'].std())
print("분산: ",df['가격'].var)
```

```python
# 1사분위수
print("분위수 25% 값", df['가격'].quantile(.25))
# 3사분위수
print("분위수 75% 값", df['가격'].quantile(.75))
```

```python
# 하위 25% 데이터
cond = df['가격'].quantile(.25) > df['가격']
df[cond]
```

```python
# 상위 25% 데이터
cond = df['가격'].quantile(.75) < df['가격']
df[cond]
```

```python
# 최빈값 구히기(여러개가 동일할수도있기떄문에 Series로 찾음)
df['원산지'].mode()[0]
```

```python
# 가격에서 가장 큰 index값 반환
df['가격'].idxmax()
```

```python
max_ind = df['가격'].idxmax()
df.loc[max_ind]
```

```python
df.loc[max_ind,"메뉴"]
```

```python
df.nlargest(3,'가격')
```

```python
df.nsmallest(3,'가격')
```

```python
# apply 적용해서 새로운 컬럼 적용하기
def cal(x):
    if x >= 100:
        return "No"
    else:
        return "Yes"

df['먹어도 될까요'] = df['칼로리'].apply(cal)
df
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': {0: '쿼카', 1: '알파카', 2: '시바견'},
    '수학': {0: 90, 1: 93, 2: 85},
    '영어': {0: 92, 1: 84, 2: 86},
    '국어': {0: 91, 1: 94, 2: 83},
})

df
```

```python
# melt로 재구조 이름을 기준으로 변경
pd.melt(df, id_vars='Name')
```

```python
pd.melt(df, id_vars='Name',value_vars=['수학','영어'])
```

```python
pd.melt(df, id_vars='Name',var_name='과목',value_name='점수')
```
