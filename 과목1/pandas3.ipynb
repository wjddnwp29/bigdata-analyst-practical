## Section9. 데이터 추가/변경

```python
import pandas as pd
```

```python
df = pd.read_csv('cafe.csv')
df.head(2)
```

```python
# 결측값으로 추가, 원두 컬럼을 만들고 결측값(NaN)으로 대입
import numpy as np
df['원산지'] = np.nan
df.head(4)
```

```python
df.tail(2)
```

```python
# 리스트 형태로 데이터(행) 추가
df.loc[7] = ['크리스마스라떼',6000,300,'한국']
```

```python
df
```

```python
# cafe2 저장
df = df.drop("시즌", axis = 0)
df.to_csv('cafe2.csv',index=False)
```

## Section10. 정렬

```python
df = pd.read_csv('cafe2.csv')
df.head(2)
```

```python
# 인덱스 기준(기본값 ascending=True(오름차순))
df.sort_index(ascending=False)
```

```python
# 값 기준(기본값 ascending=True(오름차순))
# 메뉴 기준 정렬
df.sort_values('메뉴',ascending=False)
```

```python
# 가격과 메뉴 기준 정렬
# 가격을 내림차순으로 정렬하다가 만약 같은값이있으면 메뉴는 오름차순으로 정렬
# 대입연산자로 넣어줘야함.
df = df.sort_values(['가격','메뉴'],ascending=[False,True])
```

```python
# 인덱스 새로 만들기 drop = True
# 새로안하면 그냥 옆에다가 생겨버림
df.reset_index(drop=True)
```

```python
df.reset_index(drop=False)
```

## Section11. 필터링

```python
# 데이터 불러오기
df = pd.read_csv('cafe2.csv')
df.head(2)
```

```python
df['칼로리'] < 50
```

```python
# 1개 조건
cond = df['칼로리'] < 50
df[cond]
```

```python
# 1개 조건
df[df['칼로리'] < 50]
```

```python
# 칼로리 50 이상
cond = df['칼로리'] < 50
df[~cond]
```

```python
# 2개 이상 일때(AND)
cond1 = df['가격'] >= 5000
cond2 = df['칼로리'] > 100
df[cond1 & cond2]
```

```python
# 2개 이상 일떄(OR)
df[cond1 | cond2]
```

```python
# 문자열 1개 조건
# 원두 == 콜롬비아
cond = (df['원산지'] == "한국")
df[cond]
```

```python
# 문자열과 숫자 조건(AND)
cond1 = (df['원산지'] == "한국")
cond2 = df['가격'] >= 6000
df[cond1 & cond2]
```

```python
df
```

```python
df['메뉴'].isin(['녹차'])
```

```python
cond = df['메뉴'].isin(['녹차'])
df[cond]
```

```python
box = ['녹차','카푸치노','카페라떼']
cond = df['메뉴'].isin(box)
df[cond]
```

## Section13. 값 변경

```python
df = pd.read_csv("cafe3.csv")
df.head()
```

```python
# 문자 변경 : 아메리카노 -> 룽고, 녹차 -> 그린티
df = df.replace("아메리카노", "룽고")
df = df.replace("녹차", "그린티")
df
```

```python
change = {'룽고':'아메리카노','그린티':'녹차'}
df = df.replace(change)
```

```python
df
```

```python
# loc로 값 변경하기
df.loc[6,"원산지"] = "대한민국"
df.tail(3)
```

```python
# 이벤트가 전체 1000으로 변경
df.loc[1:2,'이벤트가'] = 1000
df.head()
```

```python
df.to_csv('cafe4.csv',index=False)
```
