## Section16. 그룹핑

```python
import pandas as pd
df = pd.read_csv("cafe4.csv")
df
```

```python
# 원산지 기준, 평균
df.groupby(['원산지']).mean(numeric_only=True)
```

```python
# 원산지와 칼로리 기준, 평균
df.groupby(['원산지','칼로리']).mean(numeric_only=True)
```

```python
# 원산지와 메뉴 기준(평균,합계)
df.groupby(['원산지','메뉴']).agg(['mean','sum'])
```

```python
# 1개 인덱스 형태로 리셋
df.groupby(['원산지','메뉴']).agg(['mean','sum']).reset_index()
```

```python
# transform
import pandas as pd

# 데이터 생성 (일부 결측값 포함)
df = pd.DataFrame({
    '과일': ['딸기', '블루베리', '딸기', '블루베리', '딸기', '블루베리', '딸기', '블루베리'],
    '가격': [1000, None, 1500, None, 2000, 2500, None, 1800]  # 결측값 포함
})
df
```

```python
price = df.groupby('과일')['가격'].transform("mean")
price
```

```python
df['가격'] = df['가격'].fillna(price)
df
```

```python
df = pd.DataFrame({
    '과일': ['딸기', '블루베리', '딸기', '블루베리', '딸기', '블루베리', '딸기', '블루베리'],
    '등급': ['B', 'B', 'A', 'A', 'A', 'A', 'B', 'B'],   # 새로운 그룹 추가
    '가격': [1000, None, 1500, None, 2000, 2500, None, 1800]  # 결측값 포함
})
df
```

```python
# unstack
import pandas as pd

coffee_data = {
    '커피종류': ['아메리카노', '아메리카노', '아메리카노', '라떼', '라떼', '라떼'],
    '컵크기': ['Small', 'Medium', 'Large', 'Small', 'Medium', 'Large'],
    '판매량': [120, 150, 200, 100, 130, 180]
}
df = pd.DataFrame(coffee_data)
df
```

```python
# groupby
g = df.groupby(['커피종류','컵크기'])['판매량'].sum()
g 
```

```python
# 컵크기를 컬럼으로 변환
g.unstack()
```

```python
# 커피종류를 컬럼으로 변환
g.unstack(0)
```

```python
# 두 번의 unstack 적용
g.unstack().unstack()
```

## Section19. 데이터 합치기

```python
import pandas as pd

# 에피타이저 메뉴
appetizer = pd.DataFrame({
    'Menu': ['Salad', 'Soup', 'Bread'],
    'Price': [5000, 3000, 2000]
})

# 메인 메뉴
main = pd.DataFrame({
    'Menu': ['Steak', 'Pasta', 'Chicken'],
    'Price': [15000, 12000, 10000]
})

print(appetizer)
print(main)
```

```python
# 두 메뉴를 수직으로 연결
full_menu = pd.concat([appetizer,main])
full_menu
```

```python
# 두 메뉴를 수직으로 연결
full_menu = pd.concat([appetizer,main],ignore_index=True)
full_menu
```

```python
# 두 메뉴를 좌우로 연결
full_menu = pd.concat([appetizer,main], axis = 1)
full_menu
```

```python
# 메뉴와 가격
price = pd.DataFrame({
    'Menu': ['Salad', 'Soup', 'Steak', 'Pasta'],
    'Price': [5000, 3000, 15000, 12000]
})

# 메뉴와 칼로리
cal = pd.DataFrame({
    'Menu': ['Soup', 'Steak', 'Pasta', 'Salad'],
    'Calories': [100, 500, 700, 150]
})

# 두 데이터프레임을 'Menu'를 기준으로 병합
menu_info = pd.merge(price, cal, on='Menu')
menu_info
```

## Section 20. 피벗테이블

```python
import pandas as pd

# 직원별 부서 및 급여 정보
data = {
    '이름': ['서아', '다인', '채아', '예담', '종현', '태헌'],
    '부서': ['개발', '기획', '개발', '기획', '개발', '기획'],
    '급여': [3000, 3200, 3100, 3300, 2900, 3100]
}
df = pd.DataFrame(data)
print("[원본 데이터]")
print(df)

# 부서별 평균 급여 계산
pt = df.pivot_table(index='부서', values='급여', aggfunc='mean')
print("\n[부서별 평균 급여]")  #\n은 줄 바꿈(새로운 줄 시작)
print(pt)
```

```python
import pandas as pd

# 직원별 부서 및 급여 정보
data = {
    '이름': ['서아', '다인', '채아', '예담', '종현', '태헌'],
    '부서': ['개발', '개발', '개발', '기획', '개발', '기획'],
    '직급': ['대리','과장','대리','과장','대리','과장'],
    '급여': [3000, 3200, 3100, 3300, 2900, 3100]
}
df = pd.DataFrame(data)
print("[원본 데이터]")
print(df)

# 부서 및 직급별 급여 합계 계산
pt = df.pivot_table(index='부서', columns = '직급', values='급여', aggfunc='sum')
print("\n[부서 및 직급별 급여 합계]")  #\n은 줄 바꿈(새로운 줄 시작)
print(pt)
```

```python
# 다양한 집계를 위한 데이터셋
import pandas as pd

df = pd.DataFrame({
    "구분": ["전자", "전자", "전자", "전자", "전자", "가전", "가전", "가전", "가전"],
    "유형": ["일반", "일반", "일반", "특수", "특수", "일반", "일반", "특수", "특수"],
    "크기": ["소형", "대형", "대형", "소형", "소형", "대형", "소형", "소형", "대형"],
    "수량": [1, 2, 2, 3, 3, 4, 5, 6, 7],
    "금액": [2, 4, 5, 5, 6, 6, 8, 9, 9]
})
print(df)
```
