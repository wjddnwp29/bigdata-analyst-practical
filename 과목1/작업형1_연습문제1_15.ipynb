### Section3.파생변수, 정렬, 인덱싱

### Section4.값 변경, 정렬, 합계

- ‘views’ 컬럼의 결측값(NaN)을 0으로 대체하시오.
- ‘views’ 컬럼을 기준으로 내림차순 정렬하시오.
- 정렬된 상태에서 상위 10개의 값을 모두 10번째로 큰 값으로 대체하시오.
- 최종 ‘views’ 컬럼의 전체 합을 구하고, 정수형으로 출력하시오.

````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")

# 결측값0으로 대체
df['views'] = df['views'].fillna(0)

# view컬럼을 기준으로 내림차순 정렬
df = df.sort_values('views', ascending = False)

# 상위 10개의 값을 모두 10번째로 큰 값으로 대체
views_min = df.iloc[:10]['views'].min()
df.iloc[:10,-1] = views_min

# 최종 views 컬럼의 전체합
print(int(df['views'].sum()))
````

### Section6. 필터링, 분산

- ‘f3’ 컬럼이 gold이면서 ‘f2’ 컬럼이 2인 데이터를 찾으시오.
- 찾은 데이터에서 ‘f1’ 컬럼의 분산을 구하시오. (반올림 후 소수 둘째 자리까지 계산)

````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")


cond1 = df['f3'] == 'gold'
cond2 = df['f2'] == 2
cond = cond1 & cond2
df = df[cond]


## 분산
result = df['f1'].var()

print(round(result,2))
````

### Section7.값 변경(연산), 필터링 절대값
- 모든 나이(age)에 1을 더하시오.
- 20대의 ‘views’ 평균과 30대의 ‘views’ 평균의 절댓값 차이를 구하시오. (반올림 후 소수 둘째 자리까지 계산)

````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")


df['age'] = df['age'] + 1
cond1 = (df['age'] >= 20) & (df['age'] < 30)
cond2 = (df['age'] >= 30) & (df['age'] < 40)

a = df[cond1]['views'].mean()
b = df[cond2]['views'].mean()

c = abs(a - b)
print(round(c,2))
````

### Section11. 슬라이싱, 사분위수, 결측치 제거

### Section12.결측치 처리, 최빈값, 데이터 개수

- 결측치가 가장 많은 두 컬럼을 찾으시오.
첫 번째로 결측치가 많은 컬럼에서 결측치가 있는 데이터(행)를 삭제하시오.
- 두 번째로 결측치가 많은 컬럼의 결측값을 그 컬럼의 최빈값으로 대체하시오.
- ‘f3’ 컬럼의 ‘gold’ 값을 가진 데이터의 수를 정수형으로 구하시오.

````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")

# 결측치가 가장 많은 두 컬럼을 찾으시오...
#print(df.isnull().sum())

df = df.dropna(subset=['f1'])
#print(df['f3'].mode())
freq = df['f3'].mode()[0]

df['f3'] = df['f3'].fillna(freq)

# sum은 True인 개수
print(sum(df['f3'] == 'gold'))

# len은 그냥 True, False 둘다 나옴
print(len(df['f3'] == 'gold'))
````

### Section14.중복데이터 제거, 값 변경, 데이터 개수

- 중복 데이터를 제거하시오. (단, 중복이 발생한 경우 뒤에 나온 행을 제거한다.)
- ‘f3’ 컬럼의 결측치는 0, 'silver'는 1, 'gold'는 2, 'vip'는 3으로 변환하시오.
- 변환된 ‘f3’ 컬럼의 총합을 정수형으로 구하시오.


````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")


df = df.drop_duplicates()

cond1 = df['f3'].isnull()
cond2 = df['f3'] == 'silver'
cond3 = df['f3'] == 'gold'
cond4 = df['f3'] == 'vip'


df[cond1] = 0
df[cond2] = 1
df[cond3] = 2
df[cond4] = 3

result = df['f3'].sum()
print(result)
````

````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")

# 중복데이터 제거
df = df.drop_duplicates()

# 값 대체
import numpy as np
dict_list = {np.nan:0, 'silver':1, 'gold':2, 'vip':3}
df['f3'] = df['f3'].map(dict_list)

# f3컬럼의 총합 (정수형으로 출력)
print(int(df['f3'].sum()))
````

````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")

# 중복데이터 제거
df = df.drop_duplicates()

# 값 대체
df['f3'] = df['f3'].fillna(0)
dict_list = {'silver':1, 'gold':2, 'vip':3}
df['f3'] = df['f3'].map(dict_list)

# f3컬럼의 총합 (정수형으로 출력)
print(int(df['f3'].sum()))
````

### Section15. 컬럼 삭제, 행 단위 합계, 필터링
- 주어진 데이터에서 문자 자료형 컬럼을 삭제하시오.
- 숫자 자료형 컬럼의 결측치를 0으로 대체하시오.
- 각 행의 합이 3000보다 큰 값의 개수를 정수로 구하시오. (각 행의 합: ‘age’ + ‘f1’ + ‘f2’ + ‘f5’ + ‘views’)

````python
import pandas as pd
# df = pd.read_csv("type1_data1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/type1_data1.csv")

cols = df.select_dtypes(exclude='object').columns
df = df[cols]

df = df.fillna(0)

df = df.T

print(sum(df.sum() > 3000))
````
